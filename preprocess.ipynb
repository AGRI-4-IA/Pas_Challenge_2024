# Preprocessing csv dataset

```python
from datasets import Dataset
import pandas as pd

# Chargement du fichier csv de l'ANSD
dataset = pd.read_csv("dataset_rendement_sup_prod.csv")

# Regrouper les données par région et culture pour déterminer les cultures dominantes par région
grouped_data = dataset.groupby(['region', 'culture'])['Value'].sum().reset_index()

# Préparer le dataset formaté
formatted_data_recommendation = []
regions = dataset['region'].unique()

for region in regions:
    # Cultures dominantes pour la région triées par production (Value)
    region_data = grouped_data[grouped_data['region'] == region].sort_values(by='Value', ascending=False)
    top_cultures = ", ".join(region_data['culture'].head(3))  # Limité aux 3 meilleures cultures

    # Exemple d'entrée avec surface hypothétique
    formatted_data_recommendation.append({
        "instruction": "Recommander les cultures adaptées pour une région donnée et une surface donnée.",
        "input": f"Région : {region}, Surface : 100 hectares",
        "output": f"Les cultures recommandées sont : {top_cultures}."
    })

# Convertir en DataFrame pour vérifier
formatted_df_recommendation = pd.DataFrame(formatted_data_recommendation)
formatted_df_recommendation.head()

formatted_df_recommendation.to_csv('new_formated_Dataset.csv')
```
